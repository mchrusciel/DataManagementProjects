```python
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
#df = pd.read_html("CSSchedule.html")

with open("CSSchedule.html") as file:
    CourseList = []
    CourseName = []
    CourseProf = []
    CourseTime = []
    dom = BeautifulSoup(file, "html.parser")
    
    for i in range(3, 136):
        
        #print(dom.select(".row")[i].contents[5].text.strip())
        CourseProf.append(dom.select(".row")[i].contents[5].text.strip())
        
    for i in range(3, 136):
        
        CourseList.append(dom.select(".row")[i].text[:75].strip())
                          

    #print(CourseList)
    for i in range(3, 136):
        #print(dom.select(".row")[i].contents[3].text.replace("\n", "").replace("       ", " ").strip())
        CourseName.append(dom.select(".row")[i].contents[3].text.strip().replace("\n", "").replace("       ", " "))

    for i in range(3, 136):
        #print(dom.select(".row")[i].contents[7].text.strip())
        CourseTime.append(dom.select(".row")[i].contents[7].text.strip())
```

```python
data = { 'Courses' : CourseList, 'Course Name': CourseName, 'Instructor' : CourseProf, 'Course Time': CourseTime}
df =  pd.DataFrame(data = data)
```

## 4. initial column (or row) clean-up
Removing the Ease-ment column (since it only cants NaN values - you can see this in the df used in problem 3), and apartment number (since a large amount of the dataset has this value as NaN, and the column does not seem necessary for the statistics we are calculating - also can be seen in problem 3).

```python
df.drop(labels=['EASE-MENT', 'APARTMENT NUMBER'], axis=1)
```

```python
df.astype({'BLOCK': 'int64'}).dtypes
df['SALE  PRICE'] = df['SALE PRICE'].str.replace(',' , '').astype(float)
df['GROSS SQUARE FEET'] = df['GROSS SQUARE FEET'].str.replace(',' , '').astype(float)
df.astype({'GROSS SQUARE FEET': 'float64'}).dtypes

df.drop(columns='SALE PRICE', inplace=True)
```

Transformations:
- Block changed to dtype int64 since it never contains floats
- Sale Price converted to float - for easier computation alter
- Gross Square feet converted to float - for easier computations later

## 8. Bin the prices of properties sold


```python
plt.hist(df['SALE  PRICE'], bins=[q10[0], q25[0], q50[0], q75[0], q90[0]] , rwidth=.9, label= ['0th Percentile', '25th Percentile', '50th Percentile', '75th Percentile', '90th Percentile'])
```

Unusual about the dataset's prices : A large amount of them are 0 or close to 0 - not to mention quite a lot are also very highly priced

This probably skews earlier calculations - both in the sense that apartments that are sold for $0 probably shouldn't be counted, and very expensive apartments offset the means and medians

To fix this, you could just discount any apartment sold for $0

```python
dfRemovePrice0 = df[df['SALE  PRICE']>0]
dfRemovePrice0.set_index(keys=dfRemovePrice0['ADDRESS'], inplace = True, drop=False)
plt1 = dfRemovePrice0.plot.scatter(x='SALE  PRICE', y ='GROSS SQUARE FEET', s=30)
```

Documentation for above:
- First, created a new dataframe only counting the sale prices above 0
- Next, set the addresses to be the index of the new dataframe
- then plotted the sale price and gross square feet

```python
dfDate =  df[['SALE  PRICE']]
df[['SALE DATE']] = df['SALE DATE'].str.split('/', expand=True)[0]
dfDate.set_index(keys=df['SALE DATE'], inplace= True, drop = False)
dfDateSum = dfDate.sum(level ='SALE DATE')
dfDateSum
```
